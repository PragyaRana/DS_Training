```python
import pandas as pd
import numpy as np
data={
    'Name':['Arjun','Priya','Rahul','Sneha','Vikram'],
    'Age':[21,22,20,23,21],
    'City':['Delhi','Mumbai','Pune','Bangalore','Bhubaneswar'],
    'Marks':[85,90,78,95,88]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df)
df['Percentage']=(df['Marks']/100)*100
print("===After adding percentage column===")
print(df)
```

```python
df.rename(columns={'Marks':'Score','City':'Location'},inplace=True)
print("===After renaming columns===")
print(df)
print(df.columns.tolist())
```

```python
# drop a column
df_no_pct = df.drop(columns=['Percentage'])
print('===After dropping percentage===')
print(df_no_pct)

# drop a new row by index
df_no_row = df.drop(index=2)
print('\n== After dropping row index 2')
print(df_no_row )
```

```python
df=pd.DataFrame({
    'Name' : ['Arjun','Priya','Rahuk','Sneha'],
    'Score' : [85,90,78,95],
    'Age' : [21,22,20,23]
})

#apply() on a single column
df['Grade']= df['Score'].apply(lambda x : 'A+' if x >= 90 else ('A' if x>= 80 else 'B'))
print(df)

#apply() on entire row (axis=1)
df['Summary']=df.apply(lambda row: f"{row['Name']} scored {row['Score']}",axis=1)
print('\n== Row-wise apply ===')
print(df['Summary'])
print('\n=====')

df['Qualifiucation'] = df['Age'].apply(lambda x: 'Graduate' if x >= 21 else ('Not graduated'))
print(df)
```

```python
#map() on a seris - replace values using a dictionary
grade_map = {'A+':'Excllent','A':'Very Good','B':'Good'}
df['Grade_Label'] = df['Grade'].map(grade_map)
print('===After map()===')
print(df[['Name','Grade','Grade_Label']])
```

```python
#GroupBy : aggregiate
print(df.groupby('Name')['Score'].sum())
```

```python
sales_data = {
    'Region':     ['North', 'South', 'North', 'East', 'South', 'East', 'North', 'South'],
    'Product':    ['Laptop', 'Phone', 'Laptop', 'Tablet', 'Laptop', 'Phone', 'Phone', 'Tablet'],
    'Sales':      [55000, 20000, 60000, 30000, 58000, 22000, 52000, 28000],
    'Units':      [10, 50, 12, 25, 11, 45, 9, 22]
}
sales_df = pd.DataFrame(sales_data )
```

```python
print('===multiple aggregiation===')
agg_result = sales_df.groupby('Region').agg(
    Total_Sales=('Sales','sum'),
    Avg_Sales=('Sales','mean'),
    Max_Sales=('Sales','max'),
    Total_Units=('Units','sum')
)
print(agg_result)
```

```python
#Merging and Joining DAtaFrames

students = pd.DataFrame({
    'StudentID' : [1,2,3,4,5],
    'Name' : ['Arjun','Priya','Rahul','Sneha','Vikram']
})

scores = pd.DataFrame({
    'StudentID' : [1,2,3,4,6],
    'MArks' : [85,90,78,95,70]
})

#InnerJoin : only matching rows
print('=== Inner Join ===')
print(pd.merge(students,scores,on='StudentID',how='inner'))

#LeftJoin : all from left,matching from right
print('=== Left Join ===')
print(pd.merge(students,scores,on='StudentID',how='left'))

#RightJoin : all from right,matching from left
print('=== Right Join ===')
print(pd.merge(students,scores,on='StudentID',how='right'))

#OuterJoin : all rows from both
print('=== Outer Join ===')
print(pd.merge(students,scores,on='StudentID',how='outer'))
```

```python
#===conaination===
batchA=pd.DataFrame({'Name':['Arjun','Priya'],'Marks':[85,90]})
batchB=pd.DataFrame({'Name':['Rahul','Sneha'],'Marks':[78,95]})

#vertival concatination (stacking rows)
combined = pd.concat([batchA,batchB],ignore_index=True)
print(combined)
```

```python
#PivotTables
pivot= sales_df.pivot_table(values='Sales',index='Region',columns='Product',aggfunc='sum',fill_value=0)
print('=== Pivot table: sales by region and product ===')
print(pivot)
```

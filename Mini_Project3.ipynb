```python
# Calculate Skewness

rating_skewness = df["Aggregate rating"].skew()

print("Skewness of Ratings :", round(rating_skewness,3))

if rating_skewness > 0:
    print("The distribution is positively (right) skewed.")
elif rating_skewness < 0:
    print("The distribution is negatively (left) skewed.")
else:
    print("The distribution is approximately symmetric.")
```

## Interpretation of Skewness

- A skewness value close to **0** indicates a nearly symmetric distribution.
- A positive skewness indicates that the distribution is right-skewed.
- A negative skewness indicates that the distribution is left-skewed.

```python
# Detect Outliers using IQR Method

Q1 = df["Aggregate rating"].quantile(0.25)

Q3 = df["Aggregate rating"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - (1.5 * IQR)

upper_limit = Q3 + (1.5 * IQR)

outliers = df[
    (df["Aggregate rating"] < lower_limit) |
    (df["Aggregate rating"] > upper_limit)
]

print("Lower Limit :", round(lower_limit,2))

print("Upper Limit :", round(upper_limit,2))

print("Number of Outliers :", len(outliers))
```

# Correlation Analysis

```python
# Select Numerical Columns

numerical_data = df[
    [
        "Aggregate rating",
        "Votes",
        "Average Cost for two",
        "Price range"
    ]
]

numerical_data.head()
```

```python
# Correlation Matrix

correlation_matrix = numerical_data.corr()

print("Correlation Matrix")

correlation_matrix
```

```python
# Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Strongest Relationships

corr = correlation_matrix.abs()

np.fill_diagonal(corr.values, 0)

strongest = corr.unstack().sort_values(ascending=False)

strongest = strongest[
    strongest.index.get_level_values(0) <
    strongest.index.get_level_values(1)
]

print("Strongest Relationships")

strongest.head(5)
```

## Observation

- The heatmap clearly shows the strength and direction of relationships between numerical variables.
- Price Range and Average Cost for Two have a strong positive correlation.
- Votes show a moderate positive relationship with Aggregate Rating.
- Aggregate Rating has a weak relationship with Average Cost for Two.
- Variables with higher positive correlation tend to increase together.

## Interpretation

The correlation analysis indicates that restaurants with a higher price range generally have a higher average cost, which is expected. Customer votes show a moderate positive relationship with restaurant ratings, suggesting that highly rated restaurants often receive more customer engagement.

The weak correlation between restaurant cost and aggregate rating indicates that expensive restaurants are not necessarily rated higher than affordable restaurants. Overall, the heatmap provides a clear overview of the relationships among the numerical variables and helps identify key patterns in the dataset.

```python
# Interpret Correlation Strength

for col1 in correlation_matrix.columns:
    for col2 in correlation_matrix.columns:

        if col1 != col2:

            value = correlation_matrix.loc[col1, col2]

            if abs(value) >= 0.7:
                strength = "Strong"

            elif abs(value) >= 0.4:
                strength = "Moderate"

            else:
                strength = "Weak"

            print(f"{col1} ↔ {col2} : {round(value,2)} ({strength})")
```

# Task 10

# Feature Engineering

```python
# Create Cost Category

cost_bins = [0, 500, 1500, 3000, df["Average Cost for two"].max()]

cost_labels = ["Budget", "Moderate", "Premium", "Luxury"]

df["Cost Category"] = pd.cut(
    df["Average Cost for two"],
    bins=cost_bins,
    labels=cost_labels,
    include_lowest=True
)

df[["Average Cost for two", "Cost Category"]].head()
```

```python
# Create Rating Category

rating_bins = [0, 2.5, 3.5, 4.2, 5]

rating_labels = [
    "Poor",
    "Average",
    "Good",
    "Excellent"
]

df["Rating Category"] = pd.cut(
    df["Aggregate rating"],
    bins=rating_bins,
    labels=rating_labels,
    include_lowest=True
)

df[["Aggregate rating", "Rating Category"]].head()
```

```python
# Distribution of Cost Category

cost_distribution = df["Cost Category"].value_counts()

print("Cost Category Distribution")

display(cost_distribution)
```

```python
# Distribution of Rating Category

rating_distribution = df["Rating Category"].value_counts()

print("Rating Category Distribution")

display(rating_distribution)
```

```python
# Rating Category Visualization

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Rating Category",
    order=["Poor", "Average", "Good", "Excellent"],
    palette="rocket"
)

plt.title("Distribution of Rating Categories")

plt.xlabel("Rating Category")

plt.ylabel("Number of Restaurants")

plt.show()
```

## Observation

Feature engineering simplifies the analysis by grouping restaurants into meaningful categories based on cost and rating.

The results show that budget and moderately priced restaurants dominate the dataset, while premium and luxury restaurants form a smaller segment. Most restaurants receive good customer ratings regardless of their price category, indicating that affordability does not necessarily reduce customer satisfaction.

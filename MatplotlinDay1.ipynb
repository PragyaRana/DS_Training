```python
#Relationship between duration and calories
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("data.csv")

# Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["Duration"], df["Calories"])

plt.title("Relationship Between Duration and Calories")
plt.xlabel("Duration (minutes)")
plt.ylabel("Calories Burned")

plt.show()

# Correlation
correlation = df["Duration"].corr(df["Calories"])
print("Correlation:", correlation)
```

```python
#Relationship between pulse and calories
plt.figure(figsize=(8,5))
plt.scatter(df["Pulse"], df["Calories"])
plt.title("Pulse vs Calories")
plt.xlabel("Pulse")
plt.ylabel("Calories")
plt.show()
```

```python
#Line Chart
import matplotlib.pyplot as plt

plt.plot(df['Duration'])
plt.title("Workout Duration")
plt.xlabel("Workout Session")
plt.ylabel("Duration")
plt.show()
```

```python
#Histogram
plt.hist(df['Duration'])
plt.title("Distribution of Duration")
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter plot
plt.scatter(df['Duration'], df['Calories'])
plt.title("Duration vs Calories")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.show()
```

```python
#Bar chart
df['Duration'].value_counts().head().plot(kind='bar')
plt.title("Most Common Workout Durations")
plt.xlabel("Duration")
plt.ylabel("Count")
plt.show()
```

```python
#Which duration occurs most frequently?
df["Duration"].mode()[0]
```

```python
#Does duration affect calories burned?
df[["Duration","Calories"]].corr()
```

```python
#Is pulse related to calories burned?
df[["Pulse","Calories"]].corr()
```

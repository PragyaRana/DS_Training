```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
#Q1. Load the dataset and display the first 10 records
print(df.head(10))
```

```python
#Q2. Find Number of Rows, Columns, Column Names, and Data Types
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
#Q3. Generate a Line Chart for Pulse
plt.figure(figsize=(10,5))
plt.plot(df['Pulse'])
plt.title("Pulse Line Chart")
plt.xlabel("Index")
plt.ylabel("Pulse")
plt.show()
```

```python
#Q5. Create a Bar Chart Showing Average Values
avg_value=df[['Duration','Pulse','Maxpulse','Calories']].mean()
avg_value.plot(kind='bar')
plt.title('Average Values')
plt.ylabel('Average')
plt.show()
```

```python
#Q6. Histogram for Duration
plt.hist(df['Duration'],bins=10)
plt.title('Histogram for Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q7. Histogram forCalories
plt.hist(df['Calories'],bins=10)
plt.title('Histogram for Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q8. Scatter Plot: Duration vs Calories
plt.scatter(df['Duration'],df['Calories'])
plt.title('Scatter Plot: Duration vs Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.show()
```

```python
#Q9. Plot Pulse and Maxpulse on Same Graph
plt.figure(figsize=(10,5))
plt.plot(df['Pulse'],label='Pulse')
plt.plot(df['Maxpulse'],label='Maxpulse')

plt.title("Pulse and Maxpulse ")

plt.xlabel("Pulse")
plt.ylabel("Maxpulse")
plt.legend()

plt.show()
```

```python
#Q10. Pie Chart of Top 5 Most Frequent Workout Durations
top5 = (
    df['Duration']
    .value_counts()
    .sort_values(ascending=False)
    .head(5)
)

print(top5)

plt.figure(figsize=(8,8))

plt.pie(
    top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Top 5 Most Frequent Workout Durations")
plt.axis('equal')  

plt.show()
```

```python
#Q11. Box Plot for Calories
plt.boxplot(df['Calories'].dropna())
plt.title('box plot Calories')
plt.ylabel('Calories')
plt.show()
```

```python
#Q12. Box Plots for All Numerical Variables
df[['Duration','Pulse','Maxpulse','Calories']].boxplot()
plt.title("All Numerical Variables")
plt.show()
```

```python
#Q13. Subplots: Histogram of Pulse and Calories
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(df['Pulse'])
plt.title("Pulse Histogram")

plt.subplot(1,2,2)
plt.hist(df['Calories'].dropna())
plt.title("Calories Histogram")

plt.tight_layout()
plt.show()
```

```python
#Q14. Scatter Plot: Pulse vs Calories
plt.scatter(df['Pulse'],df['Calories'])
plt.title('Scatter Plot: Pulse vs Calories')

plt.xlabel('Calories')
plt.ylabel('Pulse')
plt.show()
```

```python
#Q15. Min, Max, Average Calories Burned
min_value = df['Calories'].min()
max_value =  df['Calories'].max()
Avg_value =  df['Calories'].mean()

print(min_value)
print(max_value)
print(Avg_value)

values = [min_value,max_value,Avg_value]
labels = ['Minimum','Maximum','Average']

plt.bar(labels,values)
plt.title('Caloris Statistics')
plt.show()
```

```python
#Q16. Most Frequent Workout Duration
most_freq = df['Duration'].mode()[0]

print("Most Frequent Duration:", most_freq)

df['Duration'].value_counts().plot(kind='bar')
plt.title("Workout Duration Frequency")
plt.xlabel("Duration")
plt.ylabel("Count")

plt.show()
```

```python
#Q17. Positive Relationship Between Duration and Calories?
plt.scatter(df['Duration'], df['Calories'])

plt.title("Duration vs Calories")
plt.xlabel("Duration")
plt.ylabel("Calories")

plt.show()

corr = df['Duration'].corr(df['Calories'])

print("Correlation:", corr)
```

```python
#Q18. Compare Pulse and Maxpulse Distributions
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(df['Pulse'])
plt.title("Pulse Histogram")

plt.subplot(1,2,2)
plt.hist(df['Maxpulse'])
plt.title("Maxpulse Histogram")

plt.show()
```

```python
#Q19. Average Calories Burned for Each Duration
avg_calories = df.groupby('Duration')['Calories'].mean()

avg_calories.plot(kind='bar')

plt.title("Average Calories by Duration")
plt.xlabel("Duration")
plt.ylabel("Average Calories")

plt.show()
```

```python
#Q20. Variable with Highest Variability
df[['Duration','Pulse','Maxpulse','Calories']].boxplot()

plt.title("Variability Comparison")
plt.show()
```

```python
#Q21. Variable with Most Outliers Using Box Plot
df[['Duration','Pulse','Maxpulse','Calories']].boxplot()
plt.show()
```

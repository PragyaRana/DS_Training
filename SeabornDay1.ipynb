```python
# 1. What is Seaborn?
# Seaborn is a Python Data Visualization library built on top of Matplotlib.
# Matplotlib = Raw Engine
# Seaborn = Luxury Car Built on Engine
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
sns.scatterplot(
    data=df,
    x= df.columns[0],
    y=df.columns[1]
)
plt.show()
```

```python
themes = ['darkgrid','whitegrid','dark','white','ticks']

for theme in themes:
    sns.set_theme(style=theme)

    plt.figure(figsize=(5, 3))

    sns.scatterplot(
        data=df,
        x= df.columns[0],
        y=df.columns[1]
     )
    
    plt.title(f"Theme: {theme}")
    plt.show()
     
    
```

```python
palettes =['deep','muted','pastel','bright','dark','colorblind']

for palette in palettes:
    sns.palplot(sns.color_palette(palette))
    plt.title(palette)
    plt.show()
```

```python
sns.boxplot(
    y=df[df.columns[0]]
)

plt.title('Box Plot')
plt.show()
```

```python
sns.violinplot(
    y=df[df.columns[0]]
)

plt.title('Voilin Plot')
plt.show()
```

```python
sns.barplot(
    data=df,
    x='Pulse',
    y='Calories',
    hue='Duration'
)
plt.title('Bar Plot')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)
plt.title('correlation Heatmap')
plt.show()
```

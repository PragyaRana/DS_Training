# Load Dataset

Let's examine:

- Number of rows

```python
# Display dataset dimensions
print("Shape:", df.shape)

# Display column names
print("\nColumns:")
print(df.columns)

# Statistical summary
df.describe()
```

# Line Chart

A line chart shows trends and changes.

Question:

How does Pulse vary across observations?

```python
# Create figure size
plt.figure(figsize=(10,5))

# Plot Pulse values
plt.plot(df["Pulse"])

# Add title
plt.title("Pulse Trend")

# Label x-axis
plt.xlabel("Record Number")

# Label y-axis
plt.ylabel("Pulse")

# Display grid
plt.grid(True)

# Show chart
plt.show()
```

# Multiple Line Plot

Comparing Pulse and Maxpulse together.

```python
# Set figure size
plt.figure(figsize=(10,5))

# Plot Pulse
plt.plot(df["Pulse"], label="Pulse")

# Plot Maxpulse
plt.plot(df["Maxpulse"], label="Max Pulse")

# Add title
plt.title("Pulse vs Max Pulse")

# Add legend
plt.legend()

# Add grid
plt.grid(True)

# Display graph
plt.show()
```

# Bar Chart

Bar charts compare quantities.

Let's compare average values of all columns.

```python
# Calculate averages
means = df.mean()

# Create figure
plt.figure(figsize=(8,5))

# Create bar chart horizontal
plt.barh(means.index, means.values)

# Create bar chart vertical
#plt.bar(means.index, means.values)

# Add title
plt.title("Average Values")

# Label y-axis
plt.ylabel("Average")

# Show chart
plt.show()
```

# Histogram

A histogram shows frequency distribution.

Let's examine Duration.

```python
# Set figure size
plt.figure(figsize=(8,5))

# Create histogram
plt.hist(df["Duration"], bins=10)

# Add title
plt.title("Duration Distribution")

# Label x-axis
plt.xlabel("Duration")

# Label y-axis
plt.ylabel("Frequency")

# Show chart
plt.show()
```

# Histogram of Calories Burned

This helps us understand how calorie values are distributed.

```python
# Create figure
plt.figure(figsize=(8,5))

# Plot histogram
plt.hist(df["Calories"], bins=15)

# Add title
plt.title("Calories Distribution")

# Label axes
plt.xlabel("Calories")
plt.ylabel("Frequency")

# Display chart
plt.show()
```

# Scatter Plot

Scatter plots show relationships between two variables.

Question:

Does longer exercise burn more calories?

```python
# Set figure size
plt.figure(figsize=(8,5))

# Create scatter plot
plt.scatter(df["Duration"], df["Calories"])

# Add title
plt.title("Duration vs Calories")

# Label axes
plt.xlabel("Duration")

plt.ylabel("Calories")

# Show chart
plt.show()
```

# Scatter Plot: Pulse vs Calories

Analyzing relationship between heart rate and calories burned.

```python
# Create figure
plt.figure(figsize=(8,5))

# Create scatter plot
plt.scatter(df["Pulse"], df["Calories"])

# Add title
plt.title("Pulse vs Calories")

# Add labels
plt.xlabel("Pulse")
plt.ylabel("Calories")

# Display chart
plt.show()
```

# Pie Chart

Pie charts show proportions.

We will categorize exercise durations.

```python
# Count duration frequencies
duration_counts = df["Duration"].value_counts().head(5)

# Create pie chart
plt.figure(figsize=(7,7))

plt.pie(
    duration_counts,
    labels=duration_counts.index,
    autopct='%1.1f%%'
)

# Add title
plt.title("Top Duration Categories")

# Display chart
plt.show()
```

# Subplots

Subplots allow multiple visualizations in a single figure.

```python
# Create figure with 2 plots
fig, axes = plt.subplots(1,2, figsize=(12,5))

# Histogram of Pulse
axes[0].hist(df["Pulse"])

# Add title
axes[0].set_title("Pulse Distribution")

# Histogram of Calories
axes[1].hist(df["Calories"])

# Add title
axes[1].set_title("Calories Distribution")

# Adjust layout
plt.tight_layout()

# Display
plt.show()
```

# Box Plot

A Box Plot (Box-and-Whisker Plot) is used to visualize:

- Minimum value
- First Quartile (Q1)
- Median (Q2)
- Third Quartile (Q3)
- Maximum value
- Outliers

It helps identify the spread of data and detect unusual observations.

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Remove missing values
calories = df["Calories"].dropna()

# Create a new figure
plt.figure(figsize=(8,5))

# Create box plot
plt.boxplot(calories)

# Add title
plt.title("Box Plot of Calories")

# Add y-axis label
plt.ylabel("Calories Burned")

# Display chart
plt.show()
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_columns = df.select_dtypes(include='number')

# Create figure
plt.figure(figsize=(12,6))

# Generate box plots for all numeric variables
plt.boxplot(
    numeric_columns.values,
    labels=numeric_columns.columns
)

# Add title
plt.title("Box Plot of All Variables")

# Add y-axis label
plt.ylabel("Values")

# Add grid
plt.grid(True)

# Display plot
plt.show()
```

Identify Outliers Numerically

```python
# Check outliers using IQR method

for col in df.select_dtypes(include='number').columns:

    # Calculate Q1
    Q1 = df[col].quantile(0.25)

    # Calculate Q3
    Q3 = df[col].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Lower limit
    lower = Q1 - 1.5 * IQR

    # Upper limit
    upper = Q3 + 1.5 * IQR

    # Count outliers
    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

# DPI (Dots Per Inch)
## What is DPI?
## DPI controls the resolution (image quality).
## Higher DPI means:
### * More pixels
### * Better quality
### * Larger file size

```python
# Create high-quality figure
plt.figure(figsize=(8,5), dpi=600)

# Plot graph
plt.plot(df["Calories"])

# Display graph
plt.show()
```
